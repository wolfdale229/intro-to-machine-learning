```python
from fastai.imports import *
from fastai.structured import *

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
```

```python
winefile = os.path.join('../datafiles/', 'winequality-red.csv')
wine_data = pd.read_csv(winefile)
```

```python
def display_data(dataframe):
    pd.option_context('display.max_rows', 100, 'display.max_columns', 100)
    display(dataframe)
```

```python
#Note pandas transpose method applied to the dataset, shape would be changed for the display function.
#Original shape below
display_data(wine_data.transpose())
print('Main dataset shape : \n{}'.format(wine_data.shape))
```

```python
# wine_data.hist(figsize=(20,15))
wine_data
```

```python
# Initail model training without spliting the dataset, the wine quality is my dependent variable
df_feature, df_label, nas = proc_df(wine_data.drop(wine_data.quality, axis=0), 'quality')
model = RandomForestClassifier().fit(df_feature, df_label)

print('Model score : \n{}'.format(model.score(df_feature, df_label)))
```

spliting the data into training and validation sets

```python
X_train = df_feature[:900].copy()
X_valid = df_feature[900:].copy()

y_train = df_label[:900].copy()
y_valid = df_label[900:].copy()
X_train.shape, X_valid.shape, y_train.shape, y_valid.shape
```

```python
#testing another model 
model = RandomForestClassifier(n_jobs=-1, n_estimators=25, max_depth=100).fit(X_train, y_train)
print('Model score : \n{}'.format(model.score(X_train, y_train)))
print('Model score : \n{}'.format(model.score(X_valid, y_valid)))
```
